```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import math
```

###a.

```python
def cosine(u):
  if abs(u)<=0.5:
    return 1+math.cos(2*math.pi*u)
  else:
    return 0

def density_kernel(x,h):
  min_val = min(x)-h
  max_val = max(x)+h
  vector_t = np.linspace(min_val, max_val, num=1000)
  t_values = []

  for t in vector_t:
    sum=0
    for observation_Xi in x:
      sum += cosine((observation_Xi-t)/h)
    t_values.append((1/(len(x)*h)) * sum)

  plt.plot(vector_t, t_values)
```

###b.

```python
math_grades = pd.read_csv("/content/grades (1).csv", usecols=["math"])
math_df = pd.DataFrame(math_grades)
math_np = math_df.to_numpy()
```

- for h=10

```python
density_kernel(math_np, 10)
```

- for h=15

```python
density_kernel(math_np, 15)
```

for h=20

```python
density_kernel(math_np, 20)
```

We assume that h=15 is the most suitable with regard to alternative options. The graph gets its peaks and smooth shape, represent the data fairly to some extent.

```python
grades = pd.read_csv("/content/grades (1).csv")
grades_df = pd.DataFrame(grades)
math_A = grades_df[grades_df.school=='A'].loc[:,"math"]
math_B = grades_df[grades_df.school=='B'].loc[:,"math"]
math_C = grades_df[grades_df.school=='C'].loc[:,"math"]

density_kernel(math_A,15)
density_kernel(math_B,15)
density_kernel(math_C,15)
```

- blue is A

Obviously from the above we can see that in general term of use we can rank the scores of the 3 school from low to high: B, A, C. More accuratley, all 3 schools has a distribution similiar to normal distribution, with different mean values and different standard deviation.
